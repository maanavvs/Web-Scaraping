# Importing the necessary Libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# Getting request from the website

```python
website=("https://www.swiggy.com/city/pune")

HEADERS = {'User-Agent': 'Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148'}

response=requests.get(website,headers=HEADERS)
```

```python
if response.status_code==200:
    print("Success!")
else:
    print("Error")
```

```python
soup=BeautifulSoup(response.content,'html.parser')
```

```python
soup
```

```python
results=soup.find_all('a',{'class':'RestaurantList__RestaurantAnchor-sc-1d3nl43-3 kcEtBq'})
```

```python
len(results)
```

```python
#Restaurant Name
results[0].find('div',{'class':'sc-beySbM cwvucc'}).get_text()
```

```python
#rating
results[0].find('span',{'class':'sc-beySbM evFhcR'}).get_text()
```

```python
restaurant_name=[]
rating=[]

for result in results:
    try:
        restaurant_name.append(result.find('div',{'class':'sc-beySbM cwvucc'}).get_text())
        rating.append(result.find('span',{'class':'sc-beySbM evFhcR'}).get_text())
    except:
        print("Not Available")
    
```

```python
swiggy=pd.DataFrame({'restaurant_name':restaurant_name,'rating':rating})
```
